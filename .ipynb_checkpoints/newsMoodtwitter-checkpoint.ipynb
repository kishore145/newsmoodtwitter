## Newsmood twitter App

Twitter application using python and tweepy to run vader analysis on last 100 tweets for news channels


    ** The overall tone of news tweets published by CBS, BBC, CNN and Newyork times is slightly negative
    ** The overall tone of news tweets published by FoxNews is slightly postive.
    ** Generally, most of the news tweets are maintaining neutral tone across the board. 
    ** BBC tops the list of news tweets with highest average negative rating of -0.11 while fox tops the list of news tweets with highest average positive rating of 0.008

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import tweepy
import json
import numpy as np
import os
import time
from config import consumer_key, consumer_secret, access_token, access_token_secret
```

```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Twitter API Keys
consumer_key = consumer_key
consumer_secret = consumer_secret
access_token = access_token
access_token_secret = access_token_secret
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Target News User Accounts
target_user = ("@BBCWorld", "@CBSNews", "@CNN", "@FoxNews", "@nytimes")

# Variables for holding sentiments
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
tweet_text = []
user_list = []
tweet_date = []

# Loop through each user
for user in target_user:

   
    # Loop through 5 pages of tweets (total 100 tweets)
    for x in range(1, 6):

        # Get all tweets from home feed
        public_tweets = api.user_timeline(user, page=x,  include_rts=1)

        # Loop through all tweets
        for tweet in public_tweets:

            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]
                        
            # Add each value to the appropriate list
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)
            
            # Append raw tweet data to the appropriate list
            user_list.append(tweet["user"]["name"])
            tweet_text.append(tweet["text"])
            tweet_date.append(tweet["created_at"])
                        

# Create a dictionary of results
user_results = {
                "Tweet Source": user_list,
                "Tweet Date" : tweet_date,
                "Tweet Text" : tweet_text,
                "Compound Score": compound_list,
                "Postive Score": positive_list,
                "Neutral Score": neutral_list,
                "Negative Score": negative_list
                }

# Create datafrane to store values
tweet_df =  pd.DataFrame(user_results)

#convert date stored as string to date timestamp
tweet_df["Tweet Date"] = pd.to_datetime(tweet_df["Tweet Date"])
    
# Sort the dataframe by source and date
tweet_df.sort_values(["Tweet Source","Tweet Date"], inplace=True)
tweet_df.reset_index(drop=True, inplace=True)
```

```python
#verifying the dataframe contains 100 tweets from each of the news channels
print(tweet_df.groupby(["Tweet Source"])["Tweet Source"].count())
tweet_df.head()
```

```python
# Export to CSV
file_name = str(time.strftime("%m-%d-%y")) + "-tweets.csv"
tweet_df.to_csv("analysis/" + file_name, mode = 'w',encoding="utf-8",index = False)
   
```

```python
#define colors and legend values
colors = ('blue', 'red', 'gold', 'green', 'c')
lgd = zip(tweet_df["Tweet Source"].unique(),colors)


```

```python
# Scatter Plot
fig, ax = plt.subplots()

# Loop through each user
for news in lgd:
    ax.scatter(np.arange(-len(tweet_df[tweet_df["Tweet Source"] == news[0]]),0, 1),
               tweet_df[tweet_df["Tweet Source"] == news[0]]["Compound Score"],
               alpha=0.75,marker="o",s=60,edgecolors = 'black',color = news[1], label = news[0])
    
    
# Incorporate the other graph properties
plt.xlim([-105, 5])
plt.xticks([-100, -80, -60, -40, -20, 0], [100, 80, 60, 40, 20, 0])
ax.set_xlabel('Tweet Time')

plt.ylim([-1.05, 1.05])
ax.set_ylabel('Tweet Polarity')

ax.set_title('Sentiment Analysis of Media Tweets (' + str(time.strftime("%m-%d-%y")) + ')')
ax.grid(True)

lgnd = ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,title = 'News Source')

# Save the figure
file_name = str(time.strftime("%m-%d-%y")) + "-scatterPlot.png"
plt.savefig("analysis/" + file_name, bbox_extra_artists=(lgnd, ), bbox_inches='tight')

plt.show()
```

```python
#creating tweet summary for each news agency
tweet_summary = tweet_df.groupby(["Tweet Source"], as_index = False)["Compound Score"].mean().round(3)
tweet_summary
```

```python
# define a function for finding optimal location for bar height values
def ylabelloc(compoundScore):
    if(compoundScore > 0):
        return compoundScore + 0.01  #if compoundscore is positive, value should be displayed above bar
    else:
        return compoundScore - 0.01  #if compoundscore is negative, value should be displayed below bar
```

```python
#set up x axis
x_axis = np.arange(len(tweet_summary))

#set plot
fig, ax = plt.subplots()

#draw bar graph
ax.bar(x_axis, tweet_summary["Compound Score"], color= colors, alpha=.75, align="center", width = 1.0, edgecolor = 'black')

#Create the ticks for our bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, tweet_summary["Tweet Source"], rotation = 90)

# Set the limits of the x axis
plt.xlim(-0.75, len(x_axis)-0.25)

# Set the limits of the y axis
plt.ylim(min(tweet_summary["Compound Score"]) - 0.02, max(tweet_summary["Compound Score"]) + 0.02)

#set x, y label and title
ax.set_xlabel('News Source')
ax.set_ylabel('Tweet Polarity')
ax.set_title('Sentiment Analysis of Media Tweets (' + str(time.strftime("%m-%d-%y")) + ')')

#Loop through each bar and label the compound score against each bar
for i in range(0, len(tweet_summary)):
    ax.text(tick_locations[i]-0.3,ylabelloc(tweet_summary["Compound Score"][i]),str(tweet_summary["Compound Score"][i]), 
            color='blue', va='center', fontweight='bold')


# Save the figure
file_name = str(time.strftime("%m-%d-%y")) + "-barPlot.png"
plt.savefig("analysis/" + file_name, bbox_inches='tight')

#Display plot
plt.show()


```
